# Data preprocessing

```python
data.duplicated().sum()
```

```python
null_col=data.columns[data.isnull().any()]
null_col
```

```python
total_cells=np.product(data.shape) # To understand the total product
total_missing=data.isnull().sum().sum()
percentage_missing=(total_missing/total_cells)*100
percentage_missing
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
for col in null_col:
    if pd.api.types.is_numeric_dtype(data[col]):
        data[col] = imputer.fit_transform(data[[col]])[:, 0]
for col in null_col:
    if not pd.api.types.is_numeric_dtype(data[col]):
        data[col].fillna(data[col].mode(dropna=True).iloc[0], inplace=True)

    if data[col].isnull().sum() / len(data) > 0.5:
        data.drop(col, axis=1, inplace=True)
```

```python
total_cells=np.product(data.shape)
total_missing=data.isnull().sum().sum()
percentage_missing=(total_missing/total_cells)*100
percentage_missing
```

# IQR method

```python
from sklearn.preprocessing import RobustScaler
numeric_data=data.select_dtypes(include=['int','float'])
q1=numeric_data.quantile(0.25)
q3=numeric_data.quantile(0.75)
iqr=q3-q1
lb=1-1.5*iqr
up=3+1.5*iqr
outliers_count=((numeric_data<lb)|(numeric_data>up)).sum()
total_count=numeric_data.count()
outliers_percentage=(outliers_count/total_count)*100

outliertohandle=outliers_percentage[outliers_percentage<5].index
datafilter=numeric_data[outliertohandle]
rb=RobustScaler()
rf=rb.fit_transform(datafilter)
rf
```

# Insights

The IQR method can help identify outliers in player attributes such as overall rating, skill ratings, age, and market value.
Outliers may represent exceptional players with unusually high or low attribute values, as well as data entry errors or anomalies in the dataset.

Analyzing the IQR of player performance metrics such as goals scored, assists, tackles, and interceptions can reveal the distribution of player contributions across different matches and seasons.
Outliers in performance metrics may indicate standout performances, exceptional gameplay moments, or statistical anomalies worthy of further investigation.

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
numeric_columns = data.select_dtypes(include=['float64','int64']).columns

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data[numeric_columns])

# Creating a new DataFrame with the scaled numerical columns
scaled_df = pd.DataFrame(scaled_data)

# You can concatenate non-numeric columns back if needed
# non_numeric_columns = data.select_dtypes(exclude=['float64', 'int64']).columns
# scaled_df = pd.concat([scaled_df, data[non_numeric_columns]], axis=1)

# Now you can use 'scaled_df' for further analysis
print(scaled_df)
```

# Insights

MinMaxScaler preserves the relationships between data points and ensures that the relative distances between points are maintained after scaling.
It is particularly useful for algorithms that rely on distance-based metrics, such as clustering or nearest neighbor algorithms.

MinMaxScaler is sensitive to outliers, as it scales the data based on the minimum and maximum values. Outliers can disproportionately affect the scaling process and may lead to distortion in the scaled data.
It's important to preprocess the data to handle outliers before applying MinMaxScaler, such as removing or winsorizing extreme values.




# Dimensionality reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)  
new_data=pd.DataFrame(pca.fit_transform(scaled_df))
new_data
```

# Insights

PCA reduces the dimensionality of the data by projecting it onto a lower-dimensional subspace spanned by the principal components.
Determine the number of principal components to retain based on the explained variance ratio or a predefined threshold.
By reducing the number of dimensions, PCA can simplify the data representation, making it easier to visualize, analyze, and interpret.

Balance the interpretability of the reduced-dimensional representation with the performance of downstream analysis tasks (e.g., clustering, classification).
PCA provides a transparent and interpretable transformation of the data, but it may not always capture complex relationships as effectively as more advanced techniques.


# Model clustering

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2,random_state=20)
predict=kmeans.fit_predict(new_data)
print('centroid:',kmeans.cluster_centers_)
print('labels:',kmeans.labels_)
```

# Model evaluation

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(new_data,predict)
score
```

# Insights

K-means clustering can group players into distinct clusters based on their attributes such as overall rating, position, skill ratings, age, and market value.
Insights into player segmentation can help identify different player archetypes, playing styles, and roles within the game.
